# What is Matplotlib?

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization in various fields, including data science, engineering, and scientific research.

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2D Line Plot

- Used for bivariate analysis  
- Can be used for both categorical vs. numerical data and numerical vs. numerical data
- Use case: Time series data

```python
# plotting a simple data
sales = [500, 490, 495, 430, 590, 643, 749, 801, 920, 940, 1120, 1002]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.plot(months,sales)
```

```python
df = pd.read_csv('./Datasets/sharma-kohli.csv') # loading dataset
df.head()
```

```python
# plotting 2d line plot using the dataframe
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])
```

**Title and Labels:** We can also add the title and labels to the axes using the `title()`, `xlabel()`, and `ylabel()`.

```python
# title and labels
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])

plt.title('Rohit Vs Virat Career Comparison')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

**Labels to Lines:** It can be unclear which line represents which player's data. To avoid confusion, we can add labels to each line in the plot.

```python
# adding labels to the lines 
plt.plot(df['index'], df['V Kohli'], label='Virat Kohli')
plt.plot(df['index'], df['RG Sharma'], label='Rohit Sharma')

plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('IPL Year')
plt.ylabel('Runs Scored')

plt.legend() # this function fetches the labels provided to plot lines
```

**Line Color:** We can also customize the line color with `color()` using color name or hex code. 

```python
# custom line color using color parameter
plt.plot(df['index'], df['V Kohli'], color='#dee612')
plt.plot(df['index'], df['RG Sharma'], color="#10ec0d")

plt.title('Rohit Vs Virat Career Comparison')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

**Line Style:** Change the style of line using the `linestyle` parameter like dashed, dotted line etc.

```python
# changing the line style 
plt.plot(df['index'], df['V Kohli'], linestyle='dashed') # dashed line
plt.plot(df['index'], df['RG Sharma'], linestyle='dotted') # dotted line

plt.title('Rohit Vs Virat Career Comparison')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

**Line Width:** Line width can also be changed using the `linewidth` parameter.

```python
# changing the line width using the `linewidth` parameter
plt.plot(df['index'], df['V Kohli'], linewidth=4)
plt.plot(df['index'], df['RG Sharma'], linewidth=2)
```

**Marker:** We can also customize the marker style using the `marker` parameter.

> What is a marker? A marker is a symbol used to represent individual data points on a plot. Markers help to highlight specific values and make it easier to identify trends or patterns in the data.

- There are different styles of markers like circle (o), square (s), diamond (d), star (*), etc.

```python
# highlighting the each individual data point using the markers
plt.plot(df['index'], df['V Kohli'], marker='o') 
plt.plot(df['index'], df['RG Sharma'], marker='d')
```

- We can also customize the marker size, color, and edge color using the `markersize`, `markerfacecolor`, and `markeredgecolor` parameters.

```python
# using markersize, markerfacecolor and markeredgecolor parameters
plt.plot(df['index'], df['V Kohli'], marker='o', markersize=10, markerfacecolor='blue', markeredgecolor='black')

plt.title('Virat Kohli IPL Performance Trend by Year')
plt.xlabel('Season')
plt.ylabel('Scored')
```

**Using all above parameters in one code with shorter syntax:**

```python
plt.plot(df['index'], df['V Kohli'], 'o-b', lw=2, ms=8, mfc='yellow', mec='black', mew=2, label='Virat Kohli')

plt.title('Virat Kohli IPL Performance Trend by Year')
plt.xlabel("Year")
plt.ylabel("Runs")

plt.legend()
```

**Limiting x-axis and y-axis:** 

- `plt.xlim([xmin, xmax])`  controls the visible range of the x-axis.
- `plt.ylim([ymin, ymax])`  controls the visible range of the y-axis.

-> Without limits matplotlib auto-adjusts to fit your data.<br>
-> With limits we decide what part of the data to zoom in/out on.

Example: let's say we want to plot the Virat and Rohit runs scored comparison between year 2010-2015

```python
# using xlim to plot only the 2010-2015 comparison
plt.plot(df['index'], df['V Kohli'], 'o-', label='Virat')
plt.plot(df['index'], df['RG Sharma'], 'o-', label='Rohit')

plt.xlim(2010,2015)
plt.title('Virat & Rohit Runs Scored Comparison (1010-2015)')

plt.xlabel('Year')
plt.ylabel('Runs Scored')

plt.legend()
```

Virats scores vary between 165 and 973. What if we want to only plot the peak years only? 

```python
# using ylim to plot the peak runs scored years only
plt.plot(df['index'], df['V Kohli'], 'o-')

plt.ylim(400,1000)
```

- xlim and ylim can also be useful in case when we have outliers in our data that hide the overall trend in plot.

Example:

```python
data = [20, 30, 40, 50, 34, 22, 102, 5000]

plt.plot(data, 'o-')
```

In above example see how an outlier hide the trend of other data points. let's adress this issue by limiting the y-axis:

```python
data = [20, 30, 40, 50, 34, 22, 102, 5000]

plt.plot(data, 'o-')

plt.ylim(0, 100)
```

Now we can clearly see the trend of other data points.

**Grid** We can also add grid lines to our plot for better readability. This can be done using the `grid()` function.

```python
plt.plot(df['index'], df['V Kohli'], 'o-', label='Virat')
plt.plot(df['index'], df['RG Sharma'], 'o-', label='Rohit')

plt.grid() # enable grids
plt.legend()
```

**plt.show:** in Jupyter Notebooks, the plot is automatically displayed without needing to call `plt.show()`. However, in standalone Python scripts or other environments, you need to explicitly call `plt.show()` to render the plot.


```python
plt.plot(df['index'], df['V Kohli'], 'o-', label='Virat')
plt.plot(df['index'], df['RG Sharma'], 'o-', label='Rohit')

plt.grid() 
plt.legend()

plt.show()
```

## Bar Chart

- A bar chart is a visual representation of categorical data using rectangular bars.
- Bivariate analysis
- Numerical vs Categorical
- Use Case: Aggregate analysis of groups

```python
# plotting bar chart
orders = [50, 34, 45, 29, 22]
markets = ['US', 'UK', 'DE', 'FR', 'IT']

plt.bar(markets,orders,color='green')

plt.title('Number of Orders by each market')
plt.xlabel('Market')
plt.ylabel('Number of Orders')
```

**Horizontal bar chart:** We can also create horizontal bar charts using the `barh()` function. we use this when numbers of categories are more and names of categories are long.


```python
# horizontal bar chart
plt.barh(markets,orders,color='green')

plt.title('Number of Orders by each market')
plt.ylabel('Market')
plt.xlabel('Number of Orders')
```

**Multiple bar chart:** 

- This is useful for comparing different groups or categories side by side.
- While there is no any function for multiple bar chart in matplotlib but we can create it by playing with the position of the bars. We can use the `bar()` function with different x positions for each group.

```python
# dataset use for ploting multiple bar chart
df = pd.read_csv('./Datasets/batsman_season_record.csv')
df
```

```python
# ploting multiple bar chart

# X-axis positions
x = np.arange(len(df['batsman']))  # [0, 1, 2, ..4]
width = 0.25  # bar width

# plot each year shifted
plt.bar(x - width, df['2015'], width, label='2015')
plt.bar(x, df['2016'], width, label='2016')
plt.bar(x + width, df['2017'], width, label='2017')

plt.xticks(x, df['batsman'])   # replace positions with batsman names
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.title("Batsman Runs (2015-2017)")
plt.legend()
plt.show()
```

**Stacked Bar Chart**

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'], bottom=df['2015'])
plt.bar(df['batsman'], df['2017'], bottom=df['2015'] + df['2016'])

plt.title('Batsman Runs (2015, 2016, 2017)')
plt.xlabel('Name')
plt.ylabel('Runs')

plt.show()
```

```python
# adding the value of runs in every bar of year
plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label=2016)
plt.bar(df['batsman'], df['2017'], bottom=df['2015'] + df['2016'], label=2017)

for i, v in enumerate(df['2015']):
    plt.text(i, v/2, v, ha='center', c='white')

for i, v in enumerate(df['2016']):
    plt.text(i, df['2015'][i] + v/2, v, ha='center', c='white')

for i, v in enumerate(df['2017']):
    plt.text(i, df['2015'][i] + df['2016'][i] + v/2, v, ha='center', c='white')

plt.title('Batsman Runs (2015, 2016, 2017)')
plt.xlabel('Name')
plt.ylabel('Runs')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
# adding the total runs scored of every batsman at the top of each bar
plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label=2016)
plt.bar(df['batsman'], df['2017'], bottom=df['2015'] + df['2016'], label=2017)

for i, v in enumerate(df['2015']):
    plt.text(i, v/2, v, ha='center', c='white')

for i, v in enumerate(df['2016']):
    plt.text(i, df['2015'][i] + v/2, v, ha='center', c='white')

for i, v in enumerate(df['2017']):
    plt.text(i, df['2015'][i] + df['2016'][i] + v/2, v, ha='center', c='white')

totals = df[['2015', '2016', '2017']].sum(axis=1)
for i, v in enumerate(totals):
    plt.text(i, v + 15, v, ha='center', weight='bold')

plt.title('Batsman Runs (2015, 2016, 2017)')
plt.xlabel('Name')
plt.ylabel('Runs')

plt.legend()
plt.tight_layout()
plt.show()
```

## Histogram

- Univariate analysis
- Numerical variable
- Use case: Distribution/Frequency of a variable

```python
import matplotlib.pyplot as plt

# Sample data
data = [7, 8, 5, 6, 8, 10, 9, 6, 7, 8, 9, 10, 6, 5, 7, 8, 9, 6, 7, 8]

# Creating histogram
plt.hist(data, bins=5, edgecolor='black')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram')

plt.show()
```

Histogram on virat runs scored:

```python
# importing data
df = pd.read_csv('./Datasets/vk.csv')

df.head()
```

```python
# plotting histogram
plt.hist(df['batsman_runs'], bins=[0, 25, 50, 75, 100, 120], edgecolor='black')

# set x-axis ticks to match the histogram bins for better readability
plt.xticks([0, 25, 50, 75, 100, 120])

plt.xlabel("Runs Range")
plt.ylabel("Frequency")
plt.title("Histogram of Virat Kohli Runs")
plt.show()
```

- With histogram we can segreate the customers age, people income, etc into different bins and see the frequency of each bin

**Logrithmic Scale**

Now, there is a problem with the histogram, like when we have larger number values in some bins and smaller number values in some bins then the larger values hide the smaller values. To avoid this we can use log scale on y-axis using `log` parm or `plt.yscale('log')`

Example:

```python
arr = np.load('./Datasets/big-array.npy')

# arr.shape -> 11949

# plotting histogram without log=True
plt.hist(arr, bins=6)

plt.show()
```

look how the smaller values are hidden by larger values, let's use log scale on y-axis:

```python
plt.hist(arr, bins=6, log=True)

# or
# plt.yscale('log')

plt.show()
```

## Pie Chart

- Univariate/Bivariate analysis
- Categorical vs Numerical
- Use Case: To find out contribution of each category in total

```python
# pie chart on simple data
data = [89, 96, 91, 75, 78]
subjects = ['Computer', 'Physics', 'Math', 'Eng', 'Urdu']

plt.pie(data, labels=subjects)

plt.show()
```

Pie chart on real data:

```python
# dataset
df = pd.read_csv('./Datasets/gayle-175.csv')  # this dataset contains the total runs scored by all scored contribute in total
df
```

```python
df = df.drop(df[df["batsman_runs"] == 0].index) # drop row with 0 runs
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')

plt.show()
```

```python
# colors
colors = ["#4E79A7", "#F28E2B", "#E15759", "#76B7B2", "#59A14F"]

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors=colors)

plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.1,0,0.1,0.1,0.1])

plt.show()
```
