# Types of Charts in Data Visualization

Here are the most common types of charts used in data visualization:

## Bar chart

- Represents categorical data with rectangular bars.
- Useful for comparing the frequency or count of different categories.
- Variables should be categorical.



## Line chart

- Displays information as a series of data points called 'markers' connected by straight line segments.
- Useful for showing trends over time.
- Variables should be numeric.



## Scatter plot

- Displays values for typically two variables for a set of data.
- Useful for determining relationships between the variables.
- Variables should be numeric.



## Box plot

- Displays the distribution of data based on a five-number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum").
- Useful for identifying outliers and understanding the spread and skewness of the data.
- Variables should be numeric.



## Histogram

- Represents the distribution of numerical data by dividing the data into bins.
- Useful for understanding the shape and spread of the data.
- Variables should be numeric.



## Stacked bar chart

- Represents categorical data with rectangular bars, where each bar is divided into segments to represent sub-groups.
- Useful for comparing the total and individual contributions of different categories.
- Variables should be categorical.



## Grouped bar chart

- Displays values for typically two variables for a set of data.
- Useful for determining relationships between the variables.
- Variables should be numeric.



## Dot plot

- Displays individual data points along a single axis.
- Useful for comparing frequencies or counts across categories.
- Variables should be categorical or discrete.



## Area chart

- Displays quantitative data visually, using a continuous line to connect data points and fill the area beneath the line.
- Useful for showing trends over time or comparing multiple groups.
- Variables should be numeric.



## Dual-axis chart

- Combines two different types of data visualizations in a single chart, using two y-axes.
- Useful for comparing two related variables with different scales.
- Variables should be numeric.



## Bubble chart

- Displays three dimensions of data in a two-dimensional chart, using bubbles of different sizes to represent the third dimension.
- Useful for visualizing relationships between three numeric variables.
- Variables should be numeric.



## Density curve

- Displays the distribution of a continuous variable by smoothing the frequency of data points.
- Useful for identifying the shape of the distribution and detecting any underlying patterns.
- Variables should be numeric.



## Violin plot

- Combines features of box plots and density plots to show the distribution of the data.
- Useful for comparing multiple groups or categories.
- Variables should be numeric.



## Heatmap

- Displays data values as colors in a two-dimensional space.
- Useful for visualizing correlations between variables or showing the intensity of data points.
- Variables can be numeric or categorical.



## Waterfall Chart

- Displays the cumulative effect of sequentially introduced positive or negative values.
- Useful for visualizing how an initial value is affected by a series of intermediate values (e.g., revenues and expenses).
- Variables can be numeric or categorical.



## Pie chart

- Displays the proportion of categories as slices of a circular pie.
- Useful for visualizing the composition of a whole and comparing parts to the whole.
- Variables should be categorical.



## Funnel chart

- Displays the stages of a process and the conversion rates between them.
- Useful for visualizing the flow of users or data through different stages.
- Variables can be numeric or categorical.



## Bullet chart

- Displays a single measure (e.g., page views) along with a target or benchmark (e.g., downloads).
- Useful for tracking performance against goals and identifying areas for improvement.
- Variables can be numeric or categorical.
